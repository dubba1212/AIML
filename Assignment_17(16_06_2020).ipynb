```
import pandas as pd 
import numpy as np
data = pd.read_fwf("https://data.princeton.edu/wws509/datasets/salary.dat",header=None)
data
```

```
new_header = data.iloc[0]
data = data[1:]
data.columns = new_header
data
```

```
data.isnull()
```

```
convert_dict ={'sx':str,'rk':str,'yr':int,'dg':str,'yd':int,'sl':int}
data = data.astype(convert_dict)
print(data.dtypes)
```

```
data.select_dtypes(exclude=['object'])
```

```
data.select_dtypes(include=['object'])
```

```
categorical  = data.select_dtypes(include=['object'])
categorical['sx'].value_counts()
```

```
dummy_set1 = pd.get_dummies(data.sx)
dummy_set1
```

```
dummy_set2 = pd.get_dummies(data.dg)
dummy_set2
```

```
new_data = pd.concat([data,dummy_set1,dummy_set2],axis=1)
new_data
```

```
data['tmp']=1
dummy_set1['tmp']=1
pd.merge(data,dummy_set1,on=['tmp'])
```

```
new_data=new_data.drop(['sx','dg'],axis=1)
new_data
```
