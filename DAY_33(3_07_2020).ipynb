```
import pandas as pd
import numpy as np
```

```
data = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/spam.csv",
                   encoding='latin-1')
data
```

```
data.shape
```

```
data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
data.shape
```

```
#split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data.v2,data.v1,test_size=0.2)
```

```
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
#CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
#creating an object for CountVectorizer
count_vect = CountVectorizer()
```

```
#Generate the word counts for the words in the document
word_count_vector = count_vect.fit(X_train)
word_count_vector.get_feature_names()
```

```
doc_array = word_count_vector.transform(X_train).toarray()
doc_array
```

```
#coverting the array into dataframe 
feature_matrix = pd.DataFrame(doc_array,
                              columns=word_count_vector.get_feature_names())
feature_matrix
```
